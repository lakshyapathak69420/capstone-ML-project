```python
def collapse(x):
    if x == 0:
        return 0
    else:
        return 1
```

```python
df1["Diabetes_012"] = df1["Diabetes_012"].apply(collapse)
```

```python
df1.rename(columns={"Diabetes_012" : "Diabetes"}, inplace=True)
df2.rename(columns={"Diabetes_binary" : "Diabetes"}, inplace=True)
df3.rename(columns={"Diabetes_binary" : "Diabetes"}, inplace=True)
```

```python
dfc = pd.concat([df1 , df2, df3], axis=0)
dfc.shape
```

```python
cols = ['HighBP', 'HighChol', 'CholCheck',  'Smoker',
       'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits',"HvyAlcoholConsump", 'AnyHealthcare']
fig , ax = plt.subplots(nrows = 3, ncols = 3, figsize = (10 , 8))
k = 0 
for i in range(3):
    for j in range(3):
        sns.countplot(data= dfc, ax = ax[i][j], x = "Diabetes", hue=cols[k], stat="percent")
        ax[i][j].set_xlabel("Diabetes")
        ax[i][j].set_xticks([0, 1], ["No", "Yes"])
        ax[i][j].legend(labels = ["No", "yes"], title = cols[k])
        k += 1

plt.tight_layout()
plt.show();
```
